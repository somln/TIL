## 데이터 제외하기 dropna

```python
import pandas as pd
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna() #전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

- axis: index or columns 
 - index: 해당하는 row를 지우고,
 - columns: 당하는 col을 지움
- how: any or all
 - any: 데이터 중에 하나라도 NaN이 있으면 지움
 -all: row 또는 col의 데이터가 모두 NaN이면 지움

```python
df.dropna(axis='index', how='any') #NaN이 하나도 있는 row 삭제 
```

```python
df.dropna(axis='columns') #how는 any가 default, NaN이 하나라도 있는 col 삭제
```
