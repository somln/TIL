# 13. 원 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
values=[30, 25, 20, 13, 10, 2]
labels=['Python', 'Java', 'Javascript', 'C#', 'c/c++', 'ETC']

plt.pie(values, labels=labels, autopct='%.lf%%', startangle=90, counterclock=False)
#그래프 시작 지점 90도, 방향은 시계방향
plt.show()
```

```python
values=[30, 25, 20, 13, 10, 2]
labels=['Python', 'Java', 'Javascript', 'C#', 'c/c++', 'ETC']
explode=[0.2, 0.1, 0, 0, 0, 0] # 간격 띄어서 그래프 그리기

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values=[30, 25, 20, 13, 10, 2]
labels=['Python', 'Java', 'Javascript', 'C#', 'c/c++', 'ETC']
explode=[0.05]*6 # 간격 띄어서 그래프 그리기

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```
